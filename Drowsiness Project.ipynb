# 4. Real Time Detections

```python
cap = cv2.VideoCapture(0) #accessing the webcam
while cap.isOpened():
    ret, frame = cap.read() #reading data from the webcam (gets a return value and a frame from the video)
    
    # Make detections 
    results = model(frame) #apply the model to frames (video input) from the webcam
    
    cv2.imshow('YOLO', np.squeeze(results.render())) #arguments: name of the output window, what is coming out (i.e. the rendered output like previous example) 
    
    if cv2.waitKey(10) & 0xFF == ord('q'): #press 'q' to quit
        break
cap.release()
cv2.destroyAllWindows()
```

# 5. Train from scratch (make our own classes)

```python
import uuid   #unique identifier to name the pictures
import os
import time
```

```python
IMAGES_PATH = os.path.join('data', 'images') #/data/images (saved image path)
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label)) #print header saying which label are we currently collecting data for
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num)) #print subheader saying which number of picture is currently being collected
        
        # Webcam feed
        ret, frame = cap.read() #cap.read() output goes into the variables ret (return value) & frame
        
        # Naming out image path (full file path of the images with its unique name)
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg') #/yolo_data/images/awake.<unique name>.jpg
        
        # Writes out image to file 
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
